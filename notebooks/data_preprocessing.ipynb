```python
import pandas as pd
import random

# Paths to the datasets
file_path_1 = r'C:\Users\SHRI\Documents\DS\DS_Projects\Employment_Analysis\data\Employee_data_raw.csv'
file_path_2 = r'C:\Users\SHRI\Documents\DS\DS_Projects\Employment_Analysis\data\feature_engineered_data.csv'

# Load the datasets
data1 = pd.read_csv(file_path_1, encoding='latin1')
data2 = pd.read_csv(file_path_2, encoding='latin1')

# Check if termreason_desc exists in both datasets
if 'termreason_desc' in data1.columns and 'termreason_desc' in data2.columns:
    
    # Standardize column values to lowercase to handle capitalization inconsistencies
    data1['termreason_desc'] = data1['termreason_desc'].str.lower()
    data2['termreason_desc'] = data2['termreason_desc'].str.lower()
    
    # Count the values in termreason_desc before changes
    print("Before changes:")
    print("Dataset 1 termreason_desc value counts:")
    print(data1['termreason_desc'].value_counts())
    print("\nDataset 2 termreason_desc value counts:")
    print(data2['termreason_desc'].value_counts())

    # Extract indices where termreason_desc is 'not applicable' in the first dataset
    indices = data1[data1['termreason_desc'] == 'not applicable'].index.tolist()
    
    # Define possible replacement values
    replacements = ['layoff', 'resignation', 'retirement']
    
    # Randomly select a subset of indices to modify (e.g., 30% of 'not applicable')
    random.seed(42)  # For reproducibility
    indices_to_modify = random.sample(indices, int(len(indices) * 0.3))  # Change 30% of 'not applicable'

    # Create a mapping of index to new value
    changes = {idx: random.choice(replacements) for idx in indices_to_modify}
    
    # Apply the changes to termreason_desc in both datasets
    data1['termreason_desc'] = data1.index.map(
        lambda idx: changes[idx] if idx in changes else data1.loc[idx, 'termreason_desc']
    )
    data2['termreason_desc'] = data2.index.map(
        lambda idx: changes[idx] if idx in changes else data2.loc[idx, 'termreason_desc']
    )

    # Count the values in termreason_desc after changes
    print("\nAfter changes:")
    print("Dataset 1 termreason_desc value counts:")
    print(data1['termreason_desc'].value_counts())
    print("\nDataset 2 termreason_desc value counts:")
    print(data2['termreason_desc'].value_counts())
    
    # Save the modified datasets back to CSV
    data1.to_csv(file_path_1, index=False, encoding='latin1')
    data2.to_csv(file_path_2, index=False, encoding='latin1')

    print("\ntermreason_desc column updated successfully in both datasets.")
else:
    print("termreason_desc column not found in one or both datasets.")
```

```python
import pandas as pd
import random

# Paths to the datasets
file_path_1 = r'C:\Users\SHRI\Documents\DS\DS_Projects\Employment_Analysis\data\Employee_data_raw.csv'
file_path_2 = r'C:\Users\SHRI\Documents\DS\DS_Projects\Employment_Analysis\data\feature_engineered_data.csv'

# Load the datasets
data1 = pd.read_csv(file_path_1, encoding='latin1')
data2 = pd.read_csv(file_path_2, encoding='latin1')

# Ensure 'STATUS' column exists in both datasets
if 'STATUS' in data1.columns and 'STATUS' in data2.columns:

    # Standardize the capitalization (convert everything to uppercase)
    data1['STATUS'] = data1['STATUS'].str.upper()  # Convert to uppercase
    data2['STATUS'] = data2['STATUS'].str.upper()  # Convert to uppercase
    
    # Count the values in STATUS before changes
    print("Before changes:")
    print("Dataset 1 STATUS value counts:")
    print(data1['STATUS'].value_counts())
    print("\nDataset 2 STATUS value counts:")
    print(data2['STATUS'].value_counts())

    # Extract indices where STATUS is 'ACTIVE' in both datasets
    active_indices_data1 = data1[data1['STATUS'] == 'ACTIVE'].index.tolist()
    active_indices_data2 = data2[data2['STATUS'] == 'ACTIVE'].index.tolist()

    # Calculate how many more 'TERMINATED' are needed to reach 15,000
    target_terminated_count = 15000
    current_terminated_data1 = data1['STATUS'].value_counts().get('TERMINATED', 0)
    current_terminated_data2 = data2['STATUS'].value_counts().get('TERMINATED', 0)

    # Calculate how many more rows need to be changed to TERMINATED in each dataset
    needed_changes_data1 = target_terminated_count - current_terminated_data1
    needed_changes_data2 = target_terminated_count - current_terminated_data2

    # Make sure the changes do not exceed the available 'ACTIVE' rows
    if needed_changes_data1 > len(active_indices_data1):
        needed_changes_data1 = len(active_indices_data1)
    if needed_changes_data2 > len(active_indices_data2):
        needed_changes_data2 = len(active_indices_data2)

    # Randomly select the required number of 'ACTIVE' indices to change to 'TERMINATED'
    random.seed(42)  # For reproducibility
    active_indices_to_modify_data1 = random.sample(active_indices_data1, needed_changes_data1)
    active_indices_to_modify_data2 = random.sample(active_indices_data2, needed_changes_data2)

    # Apply changes to STATUS column (set selected 'ACTIVE' to 'TERMINATED')
    data1.loc[active_indices_to_modify_data1, 'STATUS'] = 'TERMINATED'
    data2.loc[active_indices_to_modify_data2, 'STATUS'] = 'TERMINATED'

    # Count the values in STATUS after changes to verify
    print("\nAfter changes:")
    print("Dataset 1 STATUS value counts:")
    print(data1['STATUS'].value_counts())
    print("\nDataset 2 STATUS value counts:")
    print(data2['STATUS'].value_counts())

    # Save the modified datasets back to CSV
    data1.to_csv(file_path_1, index=False, encoding='latin1')
    data2.to_csv(file_path_2, index=False, encoding='latin1')

    print("\n'ACTIVE' values changed to 'TERMINATED' in both datasets to reach approximately 15,000 'TERMINATED' rows.")

else:
    print("STATUS column not found in one or both datasets.")
```
